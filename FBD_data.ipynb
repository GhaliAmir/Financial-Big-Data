```python
import pandas as pd
import pickle
import yfinance as yf
```

```python
tickers = []

#This method extracts the tickers of companies in the Russell2000-index from a text document and outputs them as a string.
def parse_Russ_2000():
    
    try:
        readFile = open('./Data/2000/2000_tickers.txt','r').read()
        splitFile = readFile.split('\n')
        for eachLine in splitFile:
            splitLine = eachLine.split(' ')
            ticker = splitLine[-1]
            tickers.append(ticker)
                
    except Exception:
        print ('failed in the main loop')
    
    clean_tickers = ""
    for ticker in tickers:
        if (ticker.endswith('.A') or ticker.endswith('.B')):
            clean_tickers += (ticker[0:-2] + " ")
        else:
            clean_tickers += (ticker + " ")
    
    with open("Data/2000/russell_tickers.pkl", "wb") as f:
        pickle.dump(clean_tickers, f)
    
    return clean_tickers
```

```python
str_tickers = parse_Russ_2000()
str_tickers
```

```python
#We download all the data related to our tickers from yahoo.finance for the last 10 years, daily
data = yf.download(tickers = str_tickers, period = "10y", interval = "1d", group_by = 'ticker', auto_adjust = False,
                    prepost = True, threads = True, proxy = None)
data
```

```python
#Save the downloaded data
data.to_pickle('./Data/2000/russell_all.pkl')
```

```python
#Import the data from the saved file (to avoid running the above each time)
data = pd.read_pickle('./Data/2000/russell_all.pkl')
data.head()
```

```python
non_empty_tickers = []
tickers = data.columns.get_level_values(0).unique()

#This function saves the data of each non-empty ticker in its own file, along with the non-empty-tickers names
def save_non_empty_tickers_individually(tickers):
    for ticker in tickers:
        non_empty_ticker = data[ticker].dropna(how="all")
        
        if non_empty_ticker.shape[0] == 0: 
            continue #Go to the next ticker so empty stocks are not saved
        
        non_empty_ticker.index = pd.to_datetime(non_empty_ticker.index)
        non_empty_ticker = non_empty_ticker.round(3)
        non_empty_tickers.append(ticker)
        non_empty_ticker.to_csv(path_or_buf='./Data/2000/Clean/Individual_stocks/' + str(ticker) + '.csv')
    
    with open("Data/2000/Clean/Non_empty_tickers.pkl", "wb") as f:
        pickle.dump(non_empty_tickers, f)
    
```

```python
save_non_empty_tickers_individually(tickers)
```

```python
non_empty_tickers = pd.read_pickle('./Data/2000/Clean/Non_empty_tickers.pkl')
```

```python
#This function transforms our data into time_series data for further manpulation
def to_time_series(data, tickers, ind, price):
    
    time_series = {}
    
    for t in tickers:
        
        time_series[t] = pd.read_csv('./Data/2000/Clean/Individual_stocks/' + str(t) + '.csv')
        time_series[t] = time_series[t].set_index([ind])
        time_series[t] = time_series[t][[price]]
        time_series[t] = time_series[t].rename(columns={price:t})
        time_series[t].index = pd.to_datetime(time_series[t].index) 
        
    return time_series
```

```python
time_series_stocks = to_time_series(data, non_empty_tickers, 'Date', 'Adj Close')
```

```python
#This function joins our stocks together into the same dataframe, indexed by the Date
def df_stocks(ref_stock, tickers, stocks):
    all_ticks = []
    temp = tickers.copy()
    temp.remove(ref_stock)
    for t in temp:
        all_ticks.append(stocks[t])
        
    df = stocks[ref_stock].join(all_ticks)
    return df
```

```python
#Here we extract the tickers with 10 years data (from 03.01.2011 to 31.12.2020)
Up_to_date_available_rows = 2517

ten_years_tickers = []
for s in time_series_stocks.values():
    if (s.shape[0] == Up_to_date_available_rows):
        for col_name in s.columns:
            ten_years_tickers.append(col_name)

with open("Data/2000/Clean/10_years_tickers.pkl", "wb") as f:
        pickle.dump(ten_years_tickers, f)

len(ten_years_tickers)
```

```python
ten_years_tickers = pd.read_pickle('Data/2000/Clean/10_years_tickers.pkl')
```

```python
stocks = df_stocks(ten_years_tickers[0], ten_years_tickers, time_series_stocks).dropna()
```

```python
#This function drops the stocks that contains negative and zero values
def drop_neg_stocks(stocks):
    for cols in stocks.columns:
        if True in (stocks[cols].values <= 0):
            stocks.drop(columns=[cols],inplace=True)    
    return stocks
```

```python
#This represents our final joined clean dataframe
drop_neg_stocks(stocks)
```
